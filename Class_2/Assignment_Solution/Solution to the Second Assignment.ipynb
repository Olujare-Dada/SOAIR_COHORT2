### Question 6

```python
#Import pandas in order to read table
import pandas as pd

#Creating an instance of the table in our notebook
data = pd.read_csv("working_with_tables_homework.csv")

#Checking out the first 5 rows of the table
data.head()
```

```python
#Creating the initials column
data["Initials"] = data["Firstname"] + data["Lastname"]

#checking the result
data.head()
```

### Question 7

```python
#Creating the revenue column
data["Revenue"] = data["sales"] - data["cost"]

#Checking the result
data.head()
```

```python
#Creating the commission column
data["Commission"] = 0.15*data["Revenue"]

#Checking the result
data.head()
```

### Question 8

```python
#Creating the fullname column
data["Fullname"] = data["Firstname"] + " " + data["Lastname"]

#Checking the result
data.head()
```

```python
#The len function is used to find length of a string. We can apply the function to the entire table using the .apply() method
data["Fullname"].apply(len)
```

```python
#Creating the name length column
data["Name_Length"] = data["Fullname"].apply(len)
```

### Question 9

```python
#The first 4 characters of employee's firstname 
first = data["Firstname"].str[0:4]

#The first 4 characters of employee's lastname
last = data["Lastname"].str[0:4]

#Combining the characters:
data["Username"] = first + "_" + last
```
